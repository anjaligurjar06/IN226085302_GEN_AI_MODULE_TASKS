# Task 1: Conceptual Understanding

### 1. What is the difference between "Love" and "love" in NLP?
Ans:  "Love" and "love" is different in NLP due its case senstive nature."Love" may appear at the beginning of a sentence or represent a proper noun, while "love" is typically used as a verb or noun. Thats why we normalize first so both become "love" and are treated the same.

### 2. What happens if stopwords are not removed?
Ans: If stopwords are not removed the model gets filled with common but uninformative words (like “the”, “is”, “and”) adding noise. This can reduce accuracy and increase computation as important words get diluted.

### 3.Provide two real-world scenarios where removing stopwords can be harmful.
Ans:
1. Sentiment analysis: Removing stopwords can drop words like “not”, flipping meaning (“not good” → “good”) and giving wrong sentiment.
2. Question answering / search queries: Stopwords like “who”, “where”, “when” carry intent—removing them can distort the query and return irrelevant results.

### 4.Explain the difference between stemming and lemmatization.
Ans: Stemming reduces words to their root form by removing suffixes, often producing non-dictionary words (e.g., "running" → "run").
Lemmatization converts words to their base or dictionary form using linguistic rules (e.g. "better" → "good"). Lemmatization is more accurate but computationally expensive, while stemming is faster but less precise.

# Task 2:Build Advanced Preprocessing Function

```python
import re
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
def preprocess_text(text):
  if not text or not isinstance(text, str):
      return [], ""
  text= text.lower()
  text= re.sub(r'http\S+|www\S+', '', text)
  text= re.sub(r'\S+@\S+', '', text)
  text= re.sub(r'\d+', '', text)
  text= re.sub(r'(.)\1{2,}', r'\1', text)
  text= re.sub(r'[^a-z\s]', '', text)
  text= re.sub(r'\s+', ' ', text).strip()
  tokens= text.split()
  cleaned_tokens = [word for word in tokens if len(word)>2 or word in {"no","not"}]
  cleaned_text = " ".join(tokens)

  return tokens, cleaned_text
```

# Task 3: Stress Testing


```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
for i, sentence in enumerate(test_sentences, 1):
    tokens, cleaned = preprocess_text(sentence)

    print(f"Sentence {i}")
    print("Original Text   :", sentence)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", cleaned)
    print("=" * 60)
```

# Task 4: Token Analytics

```python
results = []
for i, sentence in enumerate(test_sentences, 1):
    tokens, cleaned = preprocess_text(sentence)
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_len = round(sum(len(t) for t in tokens) / total_tokens, 2) if total_tokens > 0 else 0
    results.append((sentence, tokens, cleaned, total_tokens, unique_tokens, avg_len))

    print(f"Sentence {i}")
    print("Original Text   :", sentence)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", cleaned)
    print("Token Count     :", total_tokens)
    print("Unique Tokens   :", unique_tokens)
    print("Avg Token Length:", avg_len)
    print("=" * 60)
```

# Task 5: Frequency Analysis

```python
from collections import Counter
all_tokens =[]
for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)
word_counts = Counter(all_tokens)
top_10 = word_counts.most_common(10)
least_5 = sorted(word_counts.items(), key=lambda x: x[1])[:5]
print("Top 10 Most Frequent Words:")
for word, count in top_10:
    print(f"{word}: {count}")
print("\nTop 5 Least Frequent Words:")
for word, count in least_5:
    print(f"{word}: {count}")
```

### Analysis
1. Data is Sparse meaning Majority of words have frequency = 1
This happen when the dataset is small.
2. Common Words Dominate
Words like "this" and "now" are frequent but not very informative

# Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(test_sentences)
print("All Tokens:")
print(result["tokens"])
print("\nClean Sentences:")
for s in result["clean_sentences"]:
    print(s)
```
